# Random Walk Model
### Stochastic processes and drift

A random walk is a mathematical object, known as a stochastic or random process, that describes a path consisting of a succession of random steps on some mathematical space. This model simulates a simple 1D random walk where a value moves up or down by 1 at each step with certain probabilities.

## How it works

At each time step, the value $V$ is updated based on a probability $p$:

| Variable | Formula | Description |
|:---|:---|:---|
| $V_{t+1}$ | $V_t + 1$ | with probability $p$ (Up) |
| $V_{t+1}$ | $V_t - 1$ | with probability $1 - p$ (Down) |

When $p = 0.5$, the walk is symmetric (unbiased). When $p \neq 0.5$, the walk has a "drift".

## Setting up the model

The Random Walk model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `value` | Initial value | 0.0 |
| `prob` | Probability of stepping up | 0.5 |
| `seed` | RNG seed for reproducibility | None |

````python
import random
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Value", "gray")
class RandomWalk(Model):
    def setup(self, value=0.0, prob=0.5, seed=None) -> None:
        self.value = value
        self.prob = prob
        self._rng = random.Random(seed)

    def execute(self) -> None:
        if self._rng.random() <= self.prob:
            self.value += 1.0
        else:
            self.value -= 1.0
````

````python
RandomWalk(value=0.0, prob=0.5, seed=42)
````

## Visualization
