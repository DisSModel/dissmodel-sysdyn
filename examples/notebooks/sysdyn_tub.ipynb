# Water-in-Tub Model
### Basic stock-and-flow dynamics

The Tub model is a simple representation of a stock (water in a tub) with a constant outflow and a periodic inflow. It demonstrates how different rates and frequencies of flows affect the accumulation of a resource.

## How it works

At each time step, the water level is updated:

| Component | Rule | Description |
|:---|:---|:---|
| Outflow | `water -= out_flow` | Constant drain each step |
| Inflow | `water += in_flow` | Added every `in_period` steps |
| Constraint | `water = max(0, water)` | The tub cannot have negative water |

## Imports

````python
env = Environment()
````

## Setting up the model

The Tub model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `water` | Initial volume (gallons) | 40.0 |
| `out_flow` | Drain per step (gallons/step) | 5.0 |
| `in_flow` | Refill amount (gallons) | 0.0 |
| `in_period` | Refill interval (steps) | 10 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Water", "blue")
class Tub(Model):
    def setup(self, water=40.0, out_flow=5.0, in_flow=0.0, in_period=10) -> None:
        self.water, self.out_flow = water, out_flow
        self.in_flow, self.in_period = in_flow, in_period
        self._step = 0

    def execute(self) -> None:
        self._step += 1
        self.water -= self.out_flow
        if self.water < 0.0: self.water = 0.0
        if self._step % self.in_period == 0: self.water += self.in_flow
````

````python
Tub(water=40, out_flow=5, in_flow=20, in_period=10)
````

## Visualization

The chart will show the water level decreasing and jumping up periodically when the refill occurs.

````python
Chart(show_legend=True, show_grid=True, title="Tub Water Level")
````

## Running the simulation

````python
env.run(100)
````

## Guided Experiments

- **Draining**: Set `in_flow=0` and watch the tub empty completely.
- **Equilibrium**: Find an `in_flow` and `in_period` that keeps the average water level stable.
- **Flooding**: Increase `in_flow` to 100. Note how the water level trends upward.
- **Frequent Refills**: Decrease `in_period` to 2.

## Conclusion

The Tub model is a fundamental example of stock-and-flow logic. It shows how the state of a system (the stock) is determined by the cumulative difference between its inflows and outflows.
