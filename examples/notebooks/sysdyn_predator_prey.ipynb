# Predator-Prey Model
### Lotka-Volterra dynamics

The Predator-Prey model (also known as the Lotka-Volterra equations) describes the dynamic interaction between two species: one that hunter (predator) and one that is hunted (prey). This interaction typically leads to coupled oscillations in both populations.

## How it works

At each time step, the populations of prey ($P$) and predators ($Q$) are updated:

| Variable | Formula | Description |
|:---|:---|:---|
| $P_{t+1}$ | $P_t + r_P \cdot P_t - d_{PP} \cdot P_t \cdot Q_t$ | Prey population update |
| $Q_{t+1}$ | $Q_t - d_Q \cdot Q_t + g_Q \cdot P_t \cdot Q_t$ | Predator population update |

Where $r_P$ is prey growth, $d_{PP}$ is the predation rate, $d_Q$ is predator death rate, and $g_Q$ is predator growth per kill.

## Setting up the model

The Predator-Prey model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `prey` | Initial prey population | 1000.0 |
| `predator` | Initial predator population | 40.0 |
| `prey_growth` | Prey intrinsic growth rate | 0.08 |
| `prey_death_pred` | Predation rate | 0.001 |
| `pred_death` | Predator intrinsic death rate | 0.02 |
| `pred_growth_kills` | Predator growth rate per kill | 0.00002 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Prey", "green")
@track_plot("Predator", "red")
class PredatorPrey(Model):
    def setup(self, predator=40.0, prey=1000.0, prey_growth=0.08, prey_death_pred=0.001, pred_death=0.02, pred_growth_kills=0.00002) -> None:
        self.predator, self.prey = predator, prey
        self.prey_growth, self.prey_death_pred = prey_growth, prey_death_pred
        self.pred_death, self.pred_growth_kills = pred_death, pred_growth_kills

    def execute(self) -> None:
        self.prey += self.prey_growth * self.prey - self.prey_death_pred * self.prey * self.predator
        self.predator += -self.pred_death * self.predator + self.pred_growth_kills * self.prey * self.predator
````

````python
PredatorPrey(prey=1000, predator=40)
````

## Visualization

The chart will show the oscillating populations of prey and predators. Note how the predator peaks follow the prey peaks with a delay.

````python
Chart(show_legend=True, show_grid=True, title="Predator-Prey Dynamics")
````
