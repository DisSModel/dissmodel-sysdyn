# Yeast Population Growth Model
### Logistic growth with carrying capacity

This model simulates yeast cell proliferation in a closed culture vessel. It uses the discrete logistic equation to show how populations grow rapidly when resources are abundant but eventually stabilize as they approach the environment's carrying capacity.

## How it works

The model updates the cell count at each step based on the growth rate and the proximity to the carrying capacity:

| Variable | Formula | Description |
|:---|:---|:---|
| $C_{t+1}$ | $C_t + C_t \cdot r \cdot (1 - C_t / K)$ | Logistic growth formula |
| $C_{t+1}$ | $\min(C_{t+1}, K)$ | Hard cap at carrying capacity |

Where $r$ is the growth rate and $K$ is the carrying capacity.

## Setting up the model

The Yeast model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `cells` | Initial number of cells | 9.6 |
| `capacity` | Maximum cell count (carrying capacity) | 665.0 |
| `rate` | Per-capita growth rate | 1.1 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Cells", "purple")
class Yeast(Model):
    """
    Yeast population growth model.
    """
    cells: float
    capacity: float
    rate: float

    def setup(self, cells: float = 9.6, capacity: float = 665.0, rate: float = 1.1) -> None:
        self.cells = cells
        self.capacity = capacity
        self.rate = rate

    def execute(self) -> None:
        self.cells += self.cells * self.rate * (1.0 - self.cells / self.capacity)
        if self.cells > self.capacity:
            self.cells = self.capacity
````

````python
Yeast(cells=9.6, capacity=665.0, rate=1.1)
````

## Visualization

The chart will display the S-shaped (sigmoid) growth curve of the yeast population as it stabilizes at the carrying capacity.

````python
Chart(show_legend=True, show_grid=True, title="Yeast Population Dynamics")
````
