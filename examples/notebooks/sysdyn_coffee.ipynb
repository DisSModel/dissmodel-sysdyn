# Coffee Cooling Model
### Newton's Law of Cooling with dissmodel

This model simulates the cooling of a hot beverage (like coffee) towards the ambient room temperature. It is a classic example of a first-order negative feedback system where the rate of change is proportional to the difference between the current state and the goal (room temperature).

## How it works

At each time step, the temperature decreases proportionally to the temperature gradient:

| Variable | Formula | Description |
|:---|:---|:---|
| $T_{t+1}$ | $T_t - k \cdot (T_t - T_{room})$ | Newton's Law of Cooling |

Where $k$ is the cooling rate constant.

## Setting up the model

The Coffee model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `temperature` | Initial beverage temperature (Â°C) | 80.0 |
| `room_temperature` | Ambient temperature (Â°C) | 20.0 |
| `cooling_rate` | Cooling speed constant $k$ | 0.1 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Temperature", "blue")
class Coffee(Model):
    """
    Newton's Law of Cooling model.
    """
    temperature: float
    room_temperature: float
    cooling_rate: float

    def setup(self, temperature: float = 80.0, room_temperature: float = 20.0, cooling_rate: float = 0.1) -> None:
        self.temperature = temperature
        self.room_temperature = room_temperature
        self.cooling_rate = cooling_rate

    def execute(self) -> None:
        self.temperature -= self.cooling_rate * (self.temperature - self.room_temperature)
````

````python
Coffee(temperature=80.0, room_temperature=20.0, cooling_rate=0.1)
````

## Visualization

The `Chart` model redraws the time series at every simulation step.

````python
Chart(show_legend=True, show_grid=True, title="Coffee Cooling")
````

## Running the simulation

- **Frigid Room**: Set `room_temperature=0`.
- **Equilibrium**: Observe how the curve flattens as it approaches room temperature.

## Conclusion

Newton's Law of Cooling shows that the rate of heat loss is proportional to the temperature difference. This principle is fundamental in thermodynamics and can be applied to everything from coffee to buildings and planetary atmospheres.
