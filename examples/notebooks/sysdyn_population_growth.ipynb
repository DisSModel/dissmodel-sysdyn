# Population Growth Model
### Exponential growth with variable rate

This model simulates a population that grows exponentially, but with a twist: the growth rate itself can change over time. This allows us to model accelerating growth (positive feedback on growth rate) or decelerating growth (negative feedback on growth rate).

## How it works

At each time step, the population and the growth rate are updated:

| Variable | Formula | Description |
|:---|:---|:---|
| $P_{t+1}$ | $P_t \cdot (1 + r_t)$ | Population update |
| $r_{t+1}$ | $r_t \cdot c$ | Growth rate update |

Where $c$ is the `growth_change` factor.

## Imports

````python
from dissmodel.core import Environment
from dissmodel_sysdyn.models import PopulationGrowth
from dissmodel.visualization import Chart
````

## ⚠️ Important: instantiation order

The `Environment` must always be created **before** any model.

```
Environment  →  PopulationGrowth  →  Chart
```

````python
env = Environment()
````

## Setting up the model

The Population Growth model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `population` | Initial population size | 60.0 |
| `growth` | Initial growth rate (0.5 = 50%) | 0.5 |
| `growth_change` | Multiplier for growth rate | 1.0 (constant) |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Population", "green")
class PopulationGrowth(Model):
    def setup(self, population=60.0, growth=0.5, growth_change=1.0) -> None:
        self.population = population
        self.growth = growth
        self.growth_change = growth_change

    def execute(self) -> None:
        self.population *= (1 + self.growth)
        self.growth *= self.growth_change
````

````python
PopulationGrowth(population=60.0, growth=0.1, growth_change=1.0)
````

## Visualization

The chart will show the exponential curve of the population over time.

````python
Chart(show_legend=True, show_grid=True, title="Population Growth")
````

## Running the simulation

````python
env.run(20)
````

## Guided Experiments

- **Constant Growth**: Run with `growth_change=1.0` (pure exponential).
- **Decelerating Growth**: Set `growth_change=0.9`. Observe how the population curve levels off.
- **Accelerating Growth**: Set `growth_change=1.05` and watch the population explode.
- **Collapse**: Try a negative `growth` or a `growth_change` that leads to a decline.

## Conclusion

Pure exponential growth is rare in nature because growth rates usually change as resources become scarce or conditions change. This model provides a simple way to explore how those changes affect the overall population trajectory.
