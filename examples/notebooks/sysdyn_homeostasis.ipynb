# Homeostasis Model
### Stable equilibrium through negative feedback

This model demonstrates how a system reaches a stable equilibrium (homeostasis) through the interplay of a constant additive gain and a rate-proportional negative feedback. It is a fundamental building block of many physiological and ecological systems.

## How it works

At each time step, a fixed amount is added to the stock, while a fraction of the stock is removed (if the rate is negative):

| Variable | Formula | Description |
|:---|:---|:---|
| $S_{t+1}$ | $S_t + g + r \cdot S_t$ | Update rule with gain and proportional feedback |

The system converges to a steady state $S^* = -g / r$.

## Setting up the model

The Homeostasis model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `stock` | Initial stock value | 0.0 |
| `gain` | Fixed increment per step | 2.0 |
| `rate` | Feedback coefficient | -0.4 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Stock", "orange")
class Homeostasis(Model):
    """
    Homeostasis model — stock with fixed inflow and proportional outflow.
    """
    stock: float
    gain: float
    rate: float

    def setup(self, stock: float = 0.0, gain: float = 2.0, rate: float = -0.4) -> None:
        self.stock = stock
        self.gain  = gain
        self.rate  = rate

    def execute(self) -> None:
        self.stock += self.gain + self.rate * self.stock
````

````python
Homeostasis(stock=0.0, gain=2.0, rate=-0.4)
````

## Visualization

The `Chart` will track the `Stock` over time as it approaches its steady state.

````python
Chart(show_legend=True, show_grid=True, title="Homeostasis")
````

## Running the simulation

````python
env.run(30)
````

## Guided Experiments

- **Vary Gain**: Change `gain` to 5.0 and observe how the steady state shifts higher ($S^* = -5.0 / -0.4 = 12.5$).
- **Vary Rate**: Change `rate` to -0.1. Note how the system takes longer to reach equilibrium.
- **Initial State**: Start with `stock=20.0`. Observe how it converges to the *same* steady state from above.
- **Positive Feedback**: Set `rate` to +0.1 and watch the stock explode (unstable growth).

## Conclusion

Homeostasis is the process by which a system maintains its internal environment in a stable state despite external changes. This model shows that a simple negative feedback loop is sufficient to create such stability.
