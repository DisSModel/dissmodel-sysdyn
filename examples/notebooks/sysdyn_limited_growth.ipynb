# Limited Growth Model
### Logistic population growth with carrying capacity

The Limited Growth model extends simple exponential growth by adding a density-dependent "brake". As the population approaches the environment's carrying capacity ($K$), the effective growth rate falls toward zero. This produces the characteristic S-shaped (sigmoidal) curve common in biological systems.

## How it works

At each time step, the population grows based on its intrinsic rate, but is constrained by the remaining "room" in the environment:

| Variable | Formula | Description |
|:---|:---|:---|
| $P_{t+1}$ | $P_t + P_t \cdot r \cdot (1 - P_t / K)$ | Discrete logistic growth update |

Where $r$ is the intrinsic growth rate and $K$ is the carrying capacity.

## Setting up the model

The Limited Growth model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `pop` | Initial population size | 300.0 |
| `rate` | Intrinsic growth rate ($r$) | 0.105 |
| `capacity` | Environmental carrying capacity ($K$) | 20,000.0 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Pop", "green")
class LimitedGrowth(Model):
    """
    Logistic population growth with a carrying capacity.
    """
    pop: float
    rate: float
    capacity: float

    def setup(self, pop: float = 300.0, rate: float = 0.105, capacity: float = 20000.0) -> None:
        self.pop      = pop
        self.rate     = rate
        self.capacity = capacity

    def execute(self) -> None:
        self.pop += self.pop * self.rate * (1.0 - self.pop / self.capacity)
````

````python
LimitedGrowth(pop=300.0, rate=0.105, capacity=20000.0)
````

## Visualization

The `Chart` will show the population size approaching the carrying capacity.

````python
Chart(show_legend=True, show_grid=True, title="Limited Population Growth")
````
