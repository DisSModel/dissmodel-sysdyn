# Chaotic Growth Model
### Deterministic chaos via the logistic map

The Chaotic Growth model represents a single-species population using the discrete logistic (quadratic) map. Despite its simplicity, this model is a classic example of how deterministic systems can exhibit complex, unpredictable behavior known as chaos.

## How it works

At each time step, the population is updated using the growth rate $r$ and the current population $P_t$:

| Variable | Formula | Description |
|:---|:---|:---|
| $P_{t+1}$ | $r \cdot P_t \cdot (1 - P_t)$ | Logistic map update rule |

The population $P$ is normalized to the interval (0, 1).

## Setting up the model

The Chaotic Growth model accepts the following parameters:

| Parameter | Description | Default |
|:---|:---|:---|
| `pop` | Initial population (0 to 1) | 0.1 |
| `rate` | Growth rate parameter $r$ | 4.0 |

````python
from dissmodel.core import Model
from dissmodel.visualization import track_plot

@track_plot("Pop", "red")
class ChaoticGrowth(Model):
    """
    Chaotic population growth via the logistic map.
    """
    pop: float
    rate: float

    def setup(self, pop: float = 0.1, rate: float = 4.0) -> None:
        self.pop  = pop
        self.rate = rate

    def execute(self) -> None:
        self.pop = self.rate * self.pop * (1.0 - self.pop)
````

````python
ChaoticGrowth(pop=0.1, rate=4.0)
````

## Visualization

The `Chart` model redraws the time series at every simulation step. It automatically tracks `pop` because the `ChaoticGrowth` class is decorated with `@track_plot`.

````python
Chart(show_legend=True, show_grid=True, title="Chaotic Growth Model")
````

## Guided Experiments

- **Stable Fixed Point**: Set `rate` to 0.5 or 1.5 and observe the convergence.
- **Oscillations**: Set `rate` to 3.2 and observe the period-2 limit cycle.
- **Chaos**: Set `rate` to 3.9 or 4.0. Note how the pattern never repeats.
- **Sensitivity**: Run two simulations with `pop=0.1` and `pop=0.10001` with `rate=4.0` and see how quickly they diverge.

## Conclusion

The logistic map demonstrates that complex behavior does not require complex rules. In real-world biological systems, growth rates can lead to stable populations, regular cycles, or seemingly random fluctuations depending on environmental conditions.
